```python
import pandas as pd
import numpy as np

def cargar_dataset(file_path):
    """
    Cargue su archivo CSV, XLSX, HTML o JSON.
    """
    if '.csv' in file_path:
        df = pd.read_csv(file_path)
    elif '.xlsx' in file_path:
        df = pd.read_excel(file_path)
    elif '.json' in file_path:
        df = pd.read_json(file_path)
    elif '.html' in file_path:
        df = pd.read_html(file_path)[0]
    else:
        raise ValueError("El archivo no se puede leer")
    return df

def fill_na_ffill(df):
    return df.ffill()

def fill_na_bfill(df):
    return df.bfill()

def fill_na_with_string(df, replacement_string):
    return df.fillna(replacement_string)

def fill_na_with_mean(df):
    return df.fillna(df.mean(numeric_only=True))

def fill_na_with_median(df):
    return df.fillna(df.median(numeric_only=True))

def fill_na_with_constant(df, constant):
    return df.fillna(constant)

def identify_na(df):
    na_por_columna = df.isna().sum()
    total_na = df.isna().sum().sum()
    return na_por_columna, total_na


if __name__ == "__main__":
    file_path = input("Escriba el t√≠tulo del archivo: ")
    try:
        data = cargar_dataset(file_path)
        data_ffill = fill_na_ffill(data)
        data_bfill = fill_na_bfill(data)
        data_string = fill_na_with_string(data, "missing")
        data_mean = fill_na_with_mean(data)
        data_median = fill_na_with_median(data)
        data_constant = fill_na_with_constant(data, 0)

        na_columns, na_total = identify_na(data)
        print("NA por columna:\n", na_columns)
        print("Total NA en DataFrame:", na_total)
        
    except ValueError:
        print("El archivo no se puede leer")
```
